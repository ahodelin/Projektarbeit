```python
import pandas as pd
import numpy as np
import csv
import matplotlib as mlp
import seaborn as sns
```

```python
# icds 2008 - 2021
version_years = pd.read_csv('/home/abel/git_repos/Projektarbeit/analysis/icd_by_year_2028_2021.csv', sep=';')
version_years = version_years[version_years['ver'] != 2007]
```

```python
version_years.head()
```

```python
mlp.pyplot.figure(figsize=(7.5,4))
sns.countplot(data=version_years, x="ver")
mlp.pyplot.title("Neue ICD-10-GM pro Jahr")
mlp.pyplot.ylabel("Anzahl")
mlp.pyplot.xlabel("Fassungen/Versionen")
mlp.pyplot.savefig("/home/abel/git_repos/Projektarbeit/figures/newicdyear.svg")
```

```python
mlp.pyplot.figure(figsize=(10,6))

sns.countplot(data=version_years, x="ver", hue="ebene")
mlp.pyplot.title("Ebene der neuen ICD-10-GM pro Jahr")
mlp.pyplot.ylabel("Anzahl")
mlp.pyplot.xlabel("Fassungen/Versionen")
mlp.pyplot.legend(labels=["3 Dreisteller", "4 Viersteller", "5 Fünfsteller"], title = "Ebene")
mlp.pyplot.savefig("/home/abel/git_repos/Projektarbeit/figures/ebeneYear.svg")
```
